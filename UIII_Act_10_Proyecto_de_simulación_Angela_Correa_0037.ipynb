### Ejercicio 14: Panadero 👨‍🍳 🍞
▶ El programa simula un "Asistente Inteligente de Panadería". El asistente no solo registra panaderos, sino que "analiza" sus especialidades y su experiencia usando estructuras condicionales (if, elif, else) para darles una categoría automática y ofrecer consejos personalizados.

```python
# --- PROGRAMA: ASISTENTE INTELIGENTE PARA GESTIÓN DE PANADEROS ---

# Al inicio, la lista de panaderos empieza vacía
lospanaderos = []

# Mensaje de bienvenida del "Asistente Virtual"
print("==================================================")
print("🤖 ¡Bienvenido al Asistente de IA para Panaderías! 🥐")
print("==================================================")
print("Hola, soy Pan-Bot. Te ayudaré a gestionar y analizar")
print("el talento de tu equipo de panaderos de forma inteligente.")

# Ciclo principal del Menú Interactivo
while True:
    print("\n--- MENÚ DE OPCIONES ---")
    print("1. Registrar y Analizar nuevo panadero")
    print("2. Ver reporte inteligente de panaderos")
    print("3. Salir del asistente")

    opcion = input("\nSelecciona una opción (1, 2 o 3): ")

    # OPCIÓN 1: Registrar y analizar (IA básica con condicionales)
    if opcion == "1":
        print("\n--- REGISTRO INTELIGENTE ---")

        # Al inicio de cada registro, el diccionario del panadero empieza vacío
        nuevo_panadero = {}

        # Captura de datos básicos
        nombre = input("Nombre del panadero: ")
        especialidad = input("Especialidad (Pan dulce / Salado / Pasteles): ").lower()
        experiencia = int(input("Años de experiencia: "))

        # Guardar los datos básicos en el diccionario
        nuevo_panadero["nombre"] = nombre
        nuevo_panadero["especialidad"] = especialidad
        nuevo_panadero["experiencia"] = experiencia

        # --- SISTEMA DE DECISIÓN AUTOMÁTICA (Simulación de IA) ---
        # El programa analiza la experiencia para asignar un rango
        if experiencia < 2:
            rango = "Principiante (Requiere supervisión)"
            consejo = "Asignar a recetas estándar y masas simples."
        elif experiencia >= 2 and experiencia <= 5:
            rango = "Intermedio (Autónomo)"
            consejo = "Puede encargarse de la producción diaria y hornos."
        else:
            rango = "Maestro Panadero (Experto)"
            consejo = "Ideal para crear nuevas recetas y capacitar a otros."

        # El programa analiza la especialidad para asignar una estación de trabajo
        if especialidad == "pan dulce":
            estacion = "Estación A (Masas dulces y conchas)"
        elif especialidad == "salado":
            estacion = "Estación B (Birotes, bolillos y baguettes)"
        elif especialidad == "pasteles":
            estacion = "Estación C (Decoración y repostería fina)"
        else:
            estacion = "Estación General (Multiusos)"

        # Guardar el análisis inteligente dentro del mismo diccionario
        nuevo_panadero["rango"] = rango
        nuevo_panadero["estacion"] = estacion
        nuevo_panadero["consejo"] = consejo

        # Agregar el diccionario completo a la lista general
        lospanaderos.append(nuevo_panadero)

        print(f"\n✅ ¡Panadero {nombre} registrado con éxito!")
        print(f"🤖 Pan-Bot determinó: Rango [{rango}] -> Asignado a [{estacion}]")

    # OPCIÓN 2: Mostrar el reporte completo usando ciclo FOR
    elif opcion == "2":
        print("\n--- REPORTE INTELIGENTE DE PERSONAL ---")

        # Verificar si la lista está vacía
        if lospanaderos == []:
            print("❌ No hay panaderos registrados en el sistema todavía.")
        else:
            print(f"Actualmente hay {len(lospanaderos)} panadero(s) en la base de datos:\n")

            # Ciclo FOR para recorrer la lista de diccionarios
            for panadero in lospanaderos:
                print(f"• Panadero: {panadero['nombre']}")
                print(f"  Especialidad: {panadero['especialidad'].capitalize()}")
                print(f"  Experiencia: {panadero['experiencia']} años")
                print(f"  🧠 Rango sugerido por IA: {panadero['rango']}")
                print(f"  📍 Estación asignada: {panadero['estacion']}")
                print(f"  💡 Consejo de Pan-Bot: {panadero['consejo']}")
                print("-" * 40) # Línea divisoria

    # OPCIÓN 3: Salir del programa
    elif opcion == "3":
        print("\n🤖 ¡Gracias por usar el Asistente de IA Pan-Bot! ¡Hasta pronto! 🥐")
        break # Rompe el ciclo 'while True' para terminar el programa

    # Condicional por si el usuario ingresa una opción inválida
    else:
        print("⚠ Opción no válida. Por favor, elige 1, 2 o 3.")
print("Programa realizado por Angela Michelle Correa Lopez NL = 14")
```
