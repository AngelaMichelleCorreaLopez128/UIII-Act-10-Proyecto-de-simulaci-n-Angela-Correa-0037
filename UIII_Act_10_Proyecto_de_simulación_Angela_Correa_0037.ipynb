### Ejercicio 14: Panadero 👨‍🍳 🍞
▶ Este programa es un **simulador de base de datos y asistente virtual**
   para gestionar panaderos. Guarda los datos de cada persona en **diccionarios dentro de una lista**. Mediante un menú interactivo, permite agregar, mostrar, buscar, modificar y eliminar registros. Además, incluye una **IA básica** que analiza el pan elegido y da un consejo de horneado automático, demostrando cómo la lógica simula respuestas inteligentes.

```python
# ============================================
# SIMULADOR DE IA PARA PANADEROS
# Lista de diccionarios
# Nivel: Preparatoria
# ============================================

# Lista vacía
panaderos = []

# Diccionario vacío
lospanaderos = {}

print("===================================")
print("     IA DE PANADEROS EXPERTOS")
print("===================================")

while True:

    print("\n========= MENÚ =========")
    print("1. Agregar panadero")
    print("2. Mostrar panaderos")
    print("3. Buscar panadero")
    print("4. Modificar panadero")
    print("5. Eliminar panadero")
    print("6. Simulación IA (Asistente de Horneado)")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # ============================================
    # AGREGAR
    # ============================================
    if opcion == "1":

        nombre = input("Nombre del panadero: ")
        especialidad = input("Especialidad (Dulce/Salado/Integral): ")
        pan_favorito = input("Pan que mejor hornea: ")
        experiencia = input("Nivel (Principiante/Intermedio/Maestro): ")

        # Diccionario
        lospanaderos = {
            "nombre": nombre,
            "especialidad": especialidad, # Se mantiene la estructura idéntica
            "pan_favorito": pan_favorito,
            "experiencia": experiencia
        }

        # Corrección interna para que guarde las llaves exactas:
        lospanaderos = {
            "nombre": nombre,
            "especialidad": especialidad,
            "pan_favorito": pan_favorito,
            "experiencia": experiencia
        }

        # Guardar en lista
        panaderos.append(lospanaderos)

        print("\nPanadero agregado correctamente.")

    # ============================================
    # MOSTRAR
    # ============================================
    elif opcion == "2":

        if len(panaderos) == 0:
            print("\nLa lista está vacía.")
        else:
            print("\n===== LISTA DE PANADEROS =====")

            contador = 1

            for datos in panaderos:

                print("\nPanadero", contador)
                print("Nombre:", datos["nombre"])
                print("Especialidad:", datos["especialidad"])
                print("Pan Favorito:", datos["pan_favorito"])
                print("Experiencia:", datos["experiencia"])

                contador = contador + 1

    # ============================================
    # BUSCAR
    # ============================================
    elif opcion == "3":

        buscar = input("Escribe el nombre del panadero a buscar: ")

        encontrado = False

        for datos in panaderos:

            if datos["nombre"] == buscar:

                print("\nPanadero encontrado")
                print("Nombre:", datos["nombre"])
                print("Especialidad:", datos["especialidad"])
                print("Pan Favorito:", datos["pan_favorito"])
                print("Experiencia:", datos["experiencia"])

                encontrado = True

        if encontrado == False:
            print("\nNo se encontró el panadero.")

    # ============================================
    # MODIFICAR
    # ============================================
    elif opcion == "4":

        modificar = input("Nombre del panadero a modificar: ")

        encontrado = False

        for datos in panaderos:

            if datos["nombre"] == modificar:

                print("\nNueva especialidad:")
                datos["especialidad"] = input()

                print("Nuevo pan favorito:")
                datos["pan_favorito"] = input()

                print("Nuevo nivel de experiencia:")
                datos["experiencia"] = input()

                print("\nDatos modificados correctamente.")

                encontrado = True

        if encontrado == False:
            print("\nNo existe ese panadero.")

    # ============================================
    # ELIMINAR
    # ============================================
    elif opcion == "5":

        eliminar = input("Nombre del panadero a eliminar: ")

        encontrado = False

        for datos in panaderos:

            if datos["nombre"] == eliminar:

                panaderos.remove(datos)

                print("\nPanadero eliminado.")

                encontrado = True

        if encontrado == False:
            print("\nNo se encontró el panadero.")

    # ============================================
    # SIMULACIÓN IA
    # ============================================
    elif opcion == "6":

        print("\n===== IA ASISTENTE DE HORNEADO =====")

        tipo = input("¿Qué tipo de pan deseas hornear hoy? (concha / bolillo / integral): ")

        if tipo == "concha":
            print("IA: Te recomiendo activar la levadura con leche tibia y usar suficiente mantequilla para la costra dulce.")

        elif tipo == "bolillo":
            print("IA: ¡Excelente! Recuerda poner una charola con agua abajo en el horno para generar vapor y crear una costra crujiente.")

        elif tipo == "integral":
            print("IA: El pan integral requiere un poco más de agua o líquido para que la masa no quede muy seca.")

        else:
            print("IA: Receta nueva detectada. ¡Asegúrate de medir bien los gramos de harina y levadura!")

    # ============================================
    # SALIR
    # ============================================
    elif opcion == "7":

        print("\nGracias por usar el sistema IA de Panaderos.")
        break

    # ============================================
    # OPCIÓN INVÁLIDA
    # ============================================
    else:
        print("\nOpción incorrecta.")
print("Programa realizado por Angela Michelle Correa Lopez NL = 14")
```
